```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from nuforc_csv_to_pandas import scrub
%matplotlib inline

#Read in UFO CSV file as a dataframe
ufo_data = scrub()

#Display headers when done loading
ufo_data.head()

#Master variable for all chart colors
theme_all = 'mako'
```

```python
#Clean/process movie data
movie_data = pd.read_csv('select_popular_alien_movies.csv', index_col=0)
movie_data.rename(columns={'Release Date':'r_date'}, inplace=True)
movie_data.r_date = pd.to_datetime(movie_data.r_date, errors='coerce')
movie_data['minus90'] = movie_data.r_date - datetime.timedelta(90)
movie_data['plus90'] = movie_data.r_date + datetime.timedelta(90)
movie_data
```

```python
#Extract date from ufo_data Date_Time.
ufo_data['Just_Date'] = ufo_data.Date_Time.dt.date
ufo_datewise = ufo_data.groupby('Just_Date')
ufo_datewise.Just_Date.count()
```

```python
x = 0
fig, ax = plt.subplots(2,5,figsize=(30,10))
sns.set_theme(palette = theme_all)
sns.set_style('whitegrid', {'axes.grid' : False})
for movie, i in movie_data.iterrows():
    x += 1
    ax = fig.add_subplot(2,5,x)
    ufo_datewise.Just_Date.count().plot.line(ax=ax)
    plt.xlim(i.minus90,i.plus90)
    plt.xlabel('Date')
    plt.ylabel('Number of Reports')
    plt.title(movie)
fig.tight_layout()
plt.grid(False)
```
